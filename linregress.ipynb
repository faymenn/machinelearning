```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Load the data
data = np.loadtxt(r"C:\Users\Fnu Aymen\Documents\Linear Regression Week 2 Data\data\ex1data1.txt", delimiter=",")

# Split into columns
X = data[:, [0]]  # First column       #City Population
y = data[:, 1]  # Second column      #Profit

# Check shapes
print("X shape:", X.shape)   
print("y shape:", y.shape)

plt.scatter(X, y)
plt.xlabel("City Population")
plt.ylabel("Profit")
plt.title("Scatter Plot of City Population vs Profit")
plt.show()
```

```python
# Load the data
data = np.loadtxt(r"C:\Users\Fnu Aymen\Documents\Linear Regression Week 2 Data\data\ex1data2.txt", delimiter=",")

# Split into columns
X2 = data[:, [0]]  # First column
y2 = data[:, 1]  # Second column

# Check shapes
print("X shape:", X2.shape)
print("y shape:", y2.shape)
```

```python
model = LinearRegression()
model.fit(X, y)

w = model.coef_[0]
b = model.intercept_

print("Intercept:", w)
print("Slope:", b)
```

```python
m = X.shape[0]
predicted = np.zeros(m)

for i in range(m):
    predicted[i] = w * X[i] + b

mse = mean_squared_error(y, predicted)
print("Mean-Squared Loss:", mse)

plt.scatter(X, y)
plt.plot(X, predicted, color='red')
plt.xlabel("City Population")
plt.ylabel("Profit")
plt.title("Linear Regression Fit")
plt.show()
```
